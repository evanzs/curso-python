## gerando histogramas com python   
# grafico mais frequente observado no dia-a-dia
# representa a distribuição de frequencia de dados
# eixo horizontal temos o limite das classes de agrupamento


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## gera 100 valores normalizados entre -3 e 3
x = np.random.randn(100)

print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
x = np.random.randn(1000)
plt.hist(x,bins=35) ## sub divide cada coluna com 35 itens
```
