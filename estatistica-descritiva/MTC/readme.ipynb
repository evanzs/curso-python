# MEDIA
- SOMO DE TODOS OS VALORES DIVIDIDO PELA QUANTIDADE DESSES VALORES
- GEOMETRICAMENTE É O CENTRO DE GRAVIDADE QUE REPRESENTA O EQUILIBRIO

# CARACTERISTICAS:
- FACIL DE ENCONTRAR
- soma de cada valor obs em relação a média é zero

# DESVANTAGEM:

- muito suscetivel, a outliers e alterações nos dados
em outras palavras a soma dos devios é zero.


## MODA

# representa o valor mais frequente entre os valores observados



## MEDIANA
# valor que ocupa a posição central de dados ordenados


## JUNÇÂO DESSAS MEDIDAS NSO DA A IDEIA DE DISPENSAO DOS DADOS

- assimetria a esquerda: media fica a esquerda sendo: media < mediana <  moda
- assimetrico a direta: media fica a direita snendo: media > mediana > moda


```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados assimétricos à esquerda
dados_esquerda = np.random.normal(loc=0, scale=1, size=1000)
dados_esquerda = -dados_esquerda ** 2  # ao quadrado para torná-lo assimétrico à esquerda

# Gerando dados simétricos
dados_simetricos = np.random.normal(loc=0, scale=1, size=1000)

# Gerando dados assimétricos à direita
dados_direita = np.random.normal(loc=0, scale=1, size=1000)
dados_direita = dados_direita ** 2  # ao quadrado para torná-lo assimétrico à direita

# Plotando os gráficos
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(dados_esquerda, bins=30, color='blue', alpha=0.7)
plt.title('Assimétrico à Esquerda')

plt.subplot(1, 3, 2)
plt.hist(dados_simetricos, bins=30, color='green', alpha=0.7)
plt.title('Simétrico')

plt.subplot(1, 3, 3)
plt.hist(dados_direita, bins=30, color='red', alpha=0.7)
plt.title('Assimétrico à Direita')

plt.tight_layout()
plt.show()
```
