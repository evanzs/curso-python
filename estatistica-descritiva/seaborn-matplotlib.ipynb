```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## GRAFICOS BARRAS OU COLUNAS

```python
y = [1,2,3,4,5,6]
x = ['n1','n2','n3','n4','n5','n6']
x2 = ['variaveis grandes 1','variaveis grandes 2','variaveis grandes 3','variaveis grandes 4','variaveis grandes 5','variaveis grandes 6']
```

```python
plt.bar(x2,y)
```

```python
## uma forma de melhorar apresentação barra na Vertical ou Horizontal, nesse caso na horiszoltar

plt.barh(x2,y)
```

## titulo e labels

```python
plt.bar(x,y,color='g')
plt.title("Grafico de Barra", size=20,)
plt.xlabel("variavel quantitiva")
plt.ylabel("categoriais")
```

## utilizando seaborn

```python
sns.barplot(x=x,y=y)
```
