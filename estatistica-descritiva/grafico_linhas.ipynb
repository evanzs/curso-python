# Grafico de linhas indicado para representar s√©ries temporais

```python
import matplotlib.pyplot as plt
import seaborn as sns

y = [1,20,3,42,5,16]
x = ['n1','n2','n3','n4','n5','n6']
x2 = ['seg','ter','qua','qui','sex']
```

```python
plt.plot(y,x)
```

```python
plt.plot(x,y,'x')
plt.xlabel('Categorias')
plt.ylabel('Quantidade')
plt.title('Titulo para grafico de linhas')
plt.show()
```

```python

sns.lineplot(x=x2,y=y)

```
