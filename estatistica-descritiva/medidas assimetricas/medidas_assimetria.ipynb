```python
df = df_flowers.iloc[:,:4]
df
```

```python
## k = (Q3 - Q1) / 2 * (P90 - P10)
dividendo  = df.quantile(0.75)  - df.quantile(0.25)
divisor = (df.quantile(0.90) - df.quantile(0.10))*2 
K = dividendo / divisor;K

# K mais proximo de 0.264 levemente concentrado do centro
# k < 0.263 , muito concentrado em torno do centro
# K > 0.263, MUITO CONCETRADO EM TORNO DO CENTRO
```
